# Ventaneado STFT

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
samples = np.arange(0,16384)
ventana = np.zeros(len(samples))

fig, axs = plt.subplots(1,3, figsize = (15,3.2))

# N/H = 2
ventanas = np.zeros((5, len(samples)))
hopsize = 256
winlen = 512
for i in range(ventanas.shape[0]):
    ventanas[i, i*hopsize:(i*hopsize+winlen)] = np.hanning(winlen)
    axs[0].plot(ventanas[i,:],  'darkorange', alpha=0.8, linewidth=1.8)
    
reconstruccion = np.sum(ventanas, axis=0)
axs[0].plot(reconstruccion, 'dodgerblue', linestyle='--', linewidth=2)
axs[0].set_xlim(-100,1700)
axs[0].set_title(r"$\frac{N}{H}= 2$", fontsize=16)
axs[0].set_ylabel('Amplitud', fontsize=16)


# N/H = 4
ventanas = np.zeros((10, len(samples)))
hopsize = 128
winlen = 512
for i in range(ventanas.shape[0]):
    ventanas[i, i*hopsize:(i*hopsize+winlen)] = np.hanning(winlen)
    axs[1].plot(ventanas[i,:],  'darkorange', alpha=0.8, linewidth=1.8)
    
reconstruccion = np.sum(ventanas, axis=0)
axs[1].plot(reconstruccion, 'dodgerblue', linestyle='--', linewidth=2, label = 'Suma')
axs[1].set_xlim(-100,1700)
axs[1].set_title(r"$\frac{N}{H}= 4$", fontsize=16)
axs[1].set_xlabel('Muestras', fontsize=16)
axs[1].plot(ventanas[1,:], 'darkorange', alpha=0.8, linewidth=1.8, label='Ventanas')
axs[1].legend(bbox_to_anchor=(0.5, 0.7), prop={'size': 14})

# N/H = 2.56
ventanas = np.zeros((7, len(samples)))
hopsize = 200
winlen = 512
for i in range(ventanas.shape[0]):
    ventanas[i, i*hopsize:(i*hopsize+winlen)] = np.hanning(winlen)
    axs[2].plot(ventanas[i,:], 'darkorange', alpha=0.8, linewidth=1.8)
reconstruccion = np.sum(ventanas, axis=0)
axs[2].plot(reconstruccion, 'dodgerblue', linestyle='--', linewidth=2)
axs[2].set_xlim(-100,1700)
axs[2].set_title(r"$\frac{N}{H}= 2.56$", fontsize=16)


plt.savefig("solapamiento.png", bbox_inches='tight')
```
